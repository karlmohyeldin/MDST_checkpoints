```python
# optional challenge - fizzbuzz
# you will need to use both iteration and control flow 
# go through all numbers from 1 to 100 in order
# if the number is a multiple of 3, print fizz
# if the number is a multiple of 5, print buzz
# if the number is a multiple of 3 and 5, print fizzbuzz and NOTHING ELSE
# if the number is neither a multiple of 3 nor a multiple of 5, print the number

for number in range(1,101):
    if number % 5 == 0 and number % 3 == 0:
        print('fizzbuzz')
    elif number % 3 == 0:
        print('fizz')
    elif number % 5 == 0:
        print('buzz')
    else:
        print(number)
```

```python
import pandas as pd
import numpy as np
```

```python
# load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial

starbies = pd.read_csv('starbucks.csv')
starbies.head()
```

```python
list(range(40,200,40))
```

```python
starbies.shape
```

```python
# output the calories, sugars, and protein columns only of every 40th row. 
starbies.loc[list(range(40, 250, 40)), ['calories', 'sugars', 'protein']]
```

```python
# select all rows with more than and including 400 calories
starbies[starbies.calories >= 400]
```

```python
# select all rows whose vitamin c content is higher than the iron content
starbies[starbies['vitamin c'] > starbies['iron']]
```

```python
# create a new column containing the caffeine per calories of each drink
starbies['caffeine_per_calorie'] = starbies.caffeine / starbies.calories
starbies[['caffeine', 'calories', 'caffeine_per_calorie']].head()
```

```python
# what is the average calorie across all items?
starbies.calories.mean()
```

```python
# how many different categories of beverages are there?
starbies.beverage.nunique()
```

```python
# what is the average # calories for each beverage category?
calories_by_beverage = starbies.groupby('beverage').calories.mean().reset_index()
calories_by_beverage.head()
```

```python
calories_by_beverage.rename(columns = {'calories': 'average calories'}, inplace = True)
calories_by_beverage.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# plot the distribution of the number of calories in drinks with a histogram
plt.hist(starbies.calories, color = 'magenta')
plt.title('Calories in Starbucks Drinks')
plt.xlabel('Calories')
plt.ylabel('Number of Drinks')
```

```python
plt.barh(calories_by_beverage['beverage'], calories_by_beverage['average calories'])
```

```python
calories_by_beverage.plot.barh('beverage', 'average calories')
```

```python
# plot calories against total fat with a scatterplot
plt.scatter(x = starbies.calories, y = starbies['total fat'])
plt.title('Calories vs Total Fat in Starbucks Drinks')
plt.xlabel('Calories')
plt.ylabel('Total Fat')
```
